Number of Missed Pickups of Red River Waste Solutions

```python
missed_pickups=trash_pickup[(trash_pickup["Request"] =="Trash - Curbside/Alley Missed Pickup") & (trash_pickup["Trash Hauler"] == 'RED RIVER')]#.count()
missed_pickups
```

Finding the Number of missed pickups for every address

```python
number_missed_pickups=missed_pickups.groupby("Incident Address")["Request"].count()
number_missed_pickups
```

 The first missed pickup will not result in a fine; however, every subsequent missed pickup will result in a $200 fine

```python
fines = number_missed_pickups.apply(lambda x: (x-1) * 200).sum()
print("Total amount of damages : $" + str(fines))
```
